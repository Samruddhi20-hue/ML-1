```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
heart_df = pd.read_csv("Heart.csv")

admission_df = pd.read_csv("admission.csv")

heart_df.head(), admission_df.head()
```

Scatter Plot for heart dataset

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=heart_df['age'], y=heart_df['chol'], hue=heart_df['sex'])
plt.title("Scatter Plot of Cholesterol vs Age")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()
```


Scatter Plot for admission dataset

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=admission_df['GRE Score'], y=admission_df['CGPA'], hue=admission_df['University Rating'])
plt.title("Scatter Plot of GRE Score vs CGPA")
plt.xlabel("GRE Score")
plt.ylabel("CGPA")
plt.show()
```

Bar plot for heart dataset

```python
plt.figure(figsize=(8,5))
sns.barplot(x=heart_df['sex'], y=heart_df['chol'], color ="orange")
plt.title("Average Cholesterol Levels by Gender")
plt.xlabel("Gender (1=Male, 0=Female)")
plt.ylabel("Cholesterol")
plt.show()
```

Bar plot for admission dataset

```python
# Strip trailing spaces from column names
admission_df.columns = admission_df.columns.str.strip()

plt.figure(figsize=(8,5))
sns.barplot(x=admission_df['University Rating'], y=admission_df['Chance of Admit'], color="aqua")
plt.title("University Rating vs Admission Chance")
plt.xlabel("University Rating")
plt.ylabel("Chance of Admit")
plt.show()
```

Box plot

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=heart_df['Sex'], y=heart_df['Chol'])
plt.title("Cholesterol Levels Distribution by Gender")
plt.xlabel("Gender (1=Male, 0=Female)")
plt.ylabel("Cholesterol")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=admission_df['University Rating'], y=admission_df['CGPA'], color="red")
plt.title("CGPA Distribution by University Rating")
plt.xlabel("University Rating")
plt.ylabel("CGPA")
plt.show()
# box plot
```

Pie chart

```python
gender_counts = heart_df['sex'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=['Male', 'Female'], autopct='%1.1f%%', colors=['blue', 'red'])
plt.title("Proportion of Heart Disease Patients by Gender")
plt.show()
```

```python
rating_counts = admission_df['University Rating'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%')
plt.title("Proportion of University Ratings")
plt.show()
# pie chart
```

Line plot

```python
plt.figure(figsize=(8,5))
sns.lineplot(x=heart_df['age'], y=heart_df['chol'], color="orange")
plt.title("Cholesterol Levels Across Different Ages")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x=admission_df['GRE Score'], y=admission_df['Chance of Admit '])  # Note the space at the end
plt.title("Admission Chances by GRE Score")
plt.xlabel("GRE Score")
plt.ylabel("Chance of Admit")
plt.show()
```
